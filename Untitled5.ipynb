```python
x=df.drop(['Outcome'],axis=1)
y=df['Outcome']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
from operator import mod
model1=LogisticRegression()
model2=SVC()
model3=RandomForestClassifier()
model4=GradientBoostingClassifier()
```

```python
colums=['LogisticRegression','SVC','RandomForestClassifier','GradientBoostingClassifier']
result1=[]
result2=[]
result3=[]
```

```python
def cal(model):
      model1.fit(x_train,y_train)
      pre=model1.predict(x_test)
      acc_result = accuracy_score(y_test, pre)
      rec_result = recall_score(y_test, pre)
      f1_result = f1_score(y_test, pre)
      result1.append(acc_result)
      result2.append(rec_result)
      result3.append(f1_result)


      con_result = confusion_matrix(y_test, pre)
      # اطبع النتائج
      print(f"Accuracy: {acc_result}")
      print(f"Recall: {rec_result}")
      print(f"F1 Score: {f1_result}")
      print(f"confusion: {con_result}")
      sns.heatmap(con_result, annot=True, fmt="d")
      print(model1)
cal(model1)
```
